```python
#Q-2
def top100ImbdMovies(url):
    titles=[]
    years=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    for i in soup.find_all('td',class_='titleColumn'):
        j=i.text.split()
        j.pop(0)
        j.pop()
        k=" ".join(j)
        titles.append(k)
    for i in soup.find_all('td',class_='titleColumn'):
        j=i.text.split()
        k=j[-1].strip('()')
        years.append(k)
    ratings=[]
    for i in soup.find_all('td',class_='ratingColumn imdbRating'):
         ratings.append(i.text.replace('\n',''))
         
    top100_ratings=ratings[:100]
    top100_titles=titles[:100]
    top100_years=years[:100]
    df=pd.DataFrame({'Movie_Name':top100_titles,'Rating':top100_ratings,'Year of Release':top100_years})
    df.index +=1
    return df
df=top100ImbdMovies('https://www.imdb.com/chart/top/')
df
```

```python
#Q-4
def Meesho(url):
    prodName=[]
    price=[]
    discounts=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    for i in soup.find_all('p',class_='NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4'):
        j=i.text
        prodName.append(j)
    for i in soup.find_all('h5',class_='hiHdyy'):
        j=i.text
        price.append(j)
   
    for i in soup.find_all('span',class_='lnonyH'):
         discounts.append(i.text.replace('off',''))
         
    
    df=pd.DataFrame({'Product_Name':prodName,'Price':price,'Discount':discounts})
    return df
df=Meesho('https://meesho.com/bagsladies/pl/p7vbp')
df
```

```python
#Q-5(a)

def top10ICCTeamMen(url):
    team=[]
    matches=[]
    points=[]
    rating=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    for i in soup.find_all('span',class_='u-hide-phablet'):
        team.append(i.text)
  #  for i in soup.find_all('td',class_='rankings-block__banner--matches'):
      #  matches.append(i.text)
    for i in soup.find_all('td',class_='u-center-text'):
        matches.append(i.text)        
    for index, item in enumerate(matches):
        if (index%2)!=0:
            points.append(item)
    for index, item in enumerate(matches):
        if (index%2)!=0:
            matches.remove(item)        
   
    for i in soup.find_all('td',class_='rankings-block__banner--rating'):
        rating.append(i.text.replace('\n','').strip(' '))
    for i in soup.find_all('td',class_='rating'):
        rating.append(i.text) 
    for i in soup.find_all('td',class_='rankings-block__banner--points'):
        points.insert(0,i.text)
    for i in soup.find_all('td',class_='rankings-block__banner--matches'):
        matches.insert(0,i.text)    
     
    top10_teams=team[:10]
    top10_matches=matches[:10]
    top10_points=points[:10]
    top10_rating=rating[:10]
   # df=pd.DataFrame({'Team':top10_teams,'Matches':top10_matches,'Points':top10_points,'Rating':top10_rating})
    df=pd.DataFrame({'Team':top10_teams,'Matches':top10_matches,'Points':top10_points,'Rating':top10_rating})
    df.index +=1
    return df
df=top10ICCTeamMen('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
df
```

```python
#Q-7
def CoreyMS(url):
    heading=[]
    date=[]
    content=[]
    youtube_link=[]
    for i in np.arange(1,18,1):
        page=requests.get(url+'page/'+str(i))
        soup=BeautifulSoup(page.content)
        for i in soup.find_all('a',class_='entry-title-link'):
            heading.append(i.text)
        for i in soup.find_all('time',class_='entry-time'):
            date.append(i.text)
        for i in soup.find_all('div',class_='entry-content'):
            content.append(i.text.strip('\n'))
        for a in soup.find_all('article'):
            videoElement=a.find('iframe',class_='youtube-player')
            if videoElement:
                youtube_link.append(videoElement['src'])
            else:
                youtube_link.append('None')
    df=pd.DataFrame({'heading':heading,'Date':date,'Content':content,'Youtube_video_link': youtube_link})
    df.index +=1
    return df
df=CoreyMS('https://coreyms.com/')
df
```

```python
#Q-9:
def dineout(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    restaurantName=[]
    cuisine=[]
    location=[]
    ratings=[]
    images=[]
    for i in soup.find_all('a',class_='restnt-name'):
        restaurantName.append(i.text)
    for i in soup.find_all('span',class_='double-line-ellipsis'):
        cuisine.append(i.text[24:])
    
    for i in soup.find_all('img',class_='no-img'):
        images.append(i['data-src'])
    for i in soup.find_all('div',class_='restnt-loc'):
        location.append(i.text)
    for i in soup.find_all('div',class_='restnt-rating rating-4'):
        ratings.append(i.text)

    df=pd.DataFrame({'Restaurant_Name':restaurantName,'Cuisine':cuisine,'Location':location,'Ratings':ratings,'Image-url':images})
    
    df.index +=1
    return df
df=dineout('https://www.dineout.co.in/delhi-restaurants/buffet-special')
df
```

```python
#Q-10:
def womenTshirts(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    prodName=[]
    price=[]
    images=[]
    for i in soup.find_all('h3'):
        prodName.append(i.text)
    for i in soup.find_all('span',class_='discountedPriceText'):
        price.append(i.text)
    for i in soup.find_all('img',class_='productImgTag'):
        images.append(i['src'])
    top10_prodName=prodName[:10]
    top10_price=price[:10]
    top10_images=images[:10]
    df=pd.DataFrame({'Name':top10_prodName,'Price':top10_price,'Image-url':top10_images})
    df.index +=1
    return df
df=womenTshirts('https://www.bewakoof.com/women-plain-t-shirts?sort=low')
df
```
